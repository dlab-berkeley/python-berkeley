```python
from datetime import date
import pandas as pd
import pandas.io.data as web

%matplotlib inline
```

```python
# Here's how you specify an arbitrary Date
date(2010, 1, 1)
```

```python
# How do we get today?
today = date.today()
today
```

```python
# So, this is the first of *this* year
beginning_of_year = date(today.year, 1, 1)
beginning_of_year
```

```python
goog_from_goog = web.DataReader("GOOG", "google", beginning_of_year, today)
```
